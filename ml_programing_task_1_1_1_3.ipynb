```python
import numpy as np

def wer(hypothesis, reference):
    # Tokenize the transcriptions into lists of words
    hypothesis_tokens = hypothesis.split()
    reference_tokens = reference.split()

    # Create a matrix to store edit distances
    d = np.zeros((len(reference_tokens) + 1, len(hypothesis_tokens) + 1))

    # Initialize the first row and column
    for i in range(len(reference_tokens) + 1):
        d[i, 0] = i
    for j in range(len(hypothesis_tokens) + 1):
        d[0, j] = j

    for i in range(1, len(reference_tokens) + 1):
        for j in range(1, len(hypothesis_tokens) + 1):

            #if the words are the same, the cost is 0, otherwise 1
            cost = 0 if reference_tokens[i - 1] == hypothesis_tokens[j - 1] else 1

            #Formula for word error rate is (insertions + deletions + substitutions)/number of words
            #This fucntion finds the shortest way to get from the reference word to the actual word
            d[i, j] = min([d[i - 1, j] + 1, d[i, j - 1] + 1, d[i - 1, j - 1] + cost])

    # Calculate WER as a percentage
    return (d[len(reference_tokens), len(hypothesis_tokens)] / len(reference_tokens)) * 100

wer_scores = []

for i in range(len(whisper_df)):
    hypothesis = whisper_df.at[i, "text"]
    reference = clean_validated_df.at[i, "text"]
    wer_score = wer(hypothesis, reference)
    wer_scores.append(wer_score)

# Calculate the overall WER as the mean (average) of all individual WERs
overall_wer = np.mean(wer_scores)

print(f"Overall WER: {overall_wer:.2f}%")

```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import string
nltk.download('punkt')

# Function to tokenize and filter out punctuation
def tokenize_and_filter(text):

    words = word_tokenize(text)

    # Filter out punctuation
    words = [word for word in words if word not in string.punctuation]

    return words

# Tokenize and filter the transcriptions into words
all_words = [word for text in clip_transcriptions for word in tokenize_and_filter(text)]

# Calculate word frequency using FreqDist
word_freq = FreqDist(all_words)

# Get the 100 most common words
most_common_words = word_freq.most_common(100)

# Calculate the overall frequency of the most common words
overall_frequency = sum([freq for word, freq in most_common_words])

# Print the 100 most common words and their frequencies
print("Top 100 Most Common Words (Excluding Punctuation):")
for word, freq in most_common_words:
    print(f"{word}: {freq}")

# Print the overall frequency of the top 100 words
print(f"Overall Frequency of Top 100 Words (Excluding Punctuation): {overall_frequency}")
```

```python
import pandas as pd
from textblob import TextBlob

# Function to get sentiment scores
def get_sentiment(text):
    analysis = TextBlob(text)
    # Sentiment polarity ranges from -1 (negative) to 1 (positive)
    # Sentiment subjectivity ranges from 0 (objective) to 1 (subjective)
    return analysis.sentiment.polarity, analysis.sentiment.subjectivity

# Create new columns for sentiment polarity and subjectivity in the DataFrame
whisper_df['sentiment_polarity'], whisper_df['sentiment_subjectivity'] = zip(*clip_transcriptions.apply(get_sentiment))

whisper_df
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize

# Function to calculate TTR for a given text
def calculate_ttr(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Calculate the total number of tokens (words)
    total_tokens = len(words)

    # Calculate the number of unique words (types)
    unique_words = len(set(words))

    # Calculate TTR (Type-Token Ratio)
    if total_tokens > 0:
        ttr = unique_words / total_tokens
    else:
        ttr = 0.0

    return ttr



# Create a new column for group numbers (assuming 10 groups)
group_size = 10
whisper_df['group'] = (whisper_df.index // group_size) + 1

# Calculate TTR for each clip and group
whisper_df['ttr'] = clip_transcriptions.apply(calculate_ttr)

# Group by 'group' column and calculate the mean TTR for each group
grouped_ttr = whisper_df.groupby('group')['ttr'].mean().reset_index()

# Print the group-wise TTR
print(grouped_ttr)
```
